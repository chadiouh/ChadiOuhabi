```python
import joblib
import pandas as pd

# Charger le modèle de régression logistique
model_optimise = joblib.load('modele_regression_logistique.pkl')

# Charger le scaler depuis le fichier
scaler = joblib.load('scaler.pkl')

# Charger le nouveau fichier de données de test
new_test_data = pd.read_csv(r'C:\Users\chouh\Desktop\Openclassrooms\Projet 12 - Détectez des faux billets avec R ou Python\billets_production_test.csv', delimiter=',')

# Appliquer le scaler aux nouvelles données de test
test_data_scaled = scaler.transform(new_test_data.drop(columns=['id']))

# Faire des prédictions
predictions = model_optimise.predict(test_data_scaled)

# Convertir les prédictions en valeurs booléennes
predictions_bool = [bool(pred) for pred in predictions]

# Ajouter les prédictions à la colonne 'id' du dataframe original
new_test_data['id'] = predictions_bool

# Afficher le dataframe mis à jour avec les prédictions
new_test_data
```

Nous avons entraîné un modèle de régression logistique sur un ensemble de 1500 billets d'euros, dont 1000 authentiques et 500 contrefaits, atteignant une précision de 99%. 

En appliquant ce modèle à un nouveau jeu de données, nous avons prédit l'authenticité de chaque billet, avec les résultats indiqués dans la colonne id. Les trois premiers billets sont prédits comme faux (False), et les deux suivants comme authentiques (True).
